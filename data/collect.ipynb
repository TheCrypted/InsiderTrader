```python
def request_lobbyists_data_with_pagination(year_interval: tuple[int, int], clients: list[str], page_size=25,filing_amount_reported_min=1000.0):
    for year in range(year_interval[0], year_interval[1] + 1):
        for client in clients:
            page = 1
            while True:
                params = {
                    "filing_year": year,
                    "page_size": page_size,
                    "page": page,
                    "filing_specific_lobbying_issues": "bill",
                    "filing_amount_reported_min": filing_amount_reported_min,
                    "client_name": client,
                }
                try:
                    response = requests.get("https://lda.senate.gov/api/v1/filings", params=params, headers={"Authorization": f"Token {API_TOKEN}"})
                    data = response.json()
                    for result in data.get("results", []):
                        result["page"] = page
                    print(f"Fetched page {page} for client {client} in year {year}, received {len(data.get('results', []))} records out of {data.get('count', 0)}.")
                except requests.RequestException as e:
                    print(f"Error fetching data: {e}")
                    break
                results = data.get("results", [])
                if not results:
                    break
                yield from results
                if data.get("next", None) is None:
                    break
                page += 1
```

```python
def request_lobbyists_client_no_entries(clients: list[str], page_size=25,filing_amount_reported_min=1000.0):
    for client in clients:
        params = {
            "page_size": page_size,
            "filing_specific_lobbying_issues": "bill",
            "filing_amount_reported_min": filing_amount_reported_min,
            "client_name": client,
        }
        try:
            response = requests.get("https://lda.senate.gov/api/v1/filings", params=params, headers={"Authorization": f"Token {API_TOKEN}"})
            data = response.json()
            count = data.get('count', 0)
            print(f"Fetched initial data for client {client}, total records: {count}.")
            if count == 0:
                yield client
        except requests.RequestException as e:
            print(f"Error fetching data: {e}")
            break
```

```python
def parse_lobbyist_activities(activity_list):
    for activity in activity_list:
        activity.pop("lobbyists", None)
        activity.pop("government_entities", None)
        activity.pop("foreign_entity_issues", None)
        description = activity.get("description", "")
        matches = set()
        for mather in [r"(H\. *R\. *\d+)", r"(S\. *\d+)", r"(S\. *J\. *Res\. *\d+)", r"(H\. *J\. *Res\. *\d+)", r"(H\. *Con\. *Res\. *\d+)", r"(S\. *Con\. *Res\. *\d+)", r"(H\. *Res\. *\d+)", r"(S\. *Res\. *\d+)"]:
            m = re.findall(mather, description)
            for match in m:
                matches.add(match)
        activity["bills"] = [match.replace(" ", "") for match in matches]
    return activity_list
```

```python
df.lobbying_activities = df.lobbying_activities.apply(parse_lobbyist_activities)
df
```

```python
new_df = df.explode("lobbying_activities", ignore_index=True)
new_df = pd.concat([new_df.drop("lobbying_activities", axis=1), new_df["lobbying_activities"].apply(lambda x : pd.Series(x ))], axis=1)
new_df["bill_id"] = new_df["bills"].apply(lambda x: [b.replace(".", "").lower() for b in x] if isinstance(x, list) and len(x) > 0 else [])
new_df
```
